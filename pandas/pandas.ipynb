## Iteracje: przeglądanie DataFrame w pętli for:
### iterrows() - zwraca wiersz w liscie
### itertuples() - zwraca wiersz w krotce

```python
# jeżeli populacja przekracza 100 milionów, to w nowej kolumnie ('Size') umieszczamy tekst 'Big'
for index, row in df_pop.iterrows():
    if row['Population'] > 100:
        df_pop.loc[index, "Size"] = ' Big'   # dopisanie kolumny Size i wartości Big dla indexów powyzej 100
        print(row['Country'], df_pop.loc[index,'Size'])
```

```python
df_pop.head() # teraz jest nowa kolumna size - uzupelniona
```

```python
# itertuples() w przeciwieństwie do iterrows(), nie zwraca ona indexu, a trzeba się do niego odwołać poprzez row.Index.
for row in df_pop.itertuples():
    if row.Population > 200:
        print(row.Index, row.Country, row.Population)
```

### FUNKCJA concat() - ZESTAWIANIE DataFrames

```python
# uzyskanie informacji o ilości krajów na region:
countries_per_region = df_pop2["Region"].value_counts()
```

```python
countries_per_region
```

```python
# uzyskanie informacji o ilości ludzi na region:
pop2_region = df_pop2.groupby(['Region'])
peoples_per_region = pop2_region['Population'].sum()
peoples_per_region
```

```python
# zestawienie dwóch powyzszych tabelek w jedną:
regions_data = pd.concat([countries_per_region, peoples_per_region], axis = 'columns', sort = False)
```

```python
regions_data
```

```python
regions_data.rename(columns={'Region': 'countries', 'Population': 'peoples'}, inplace = True)
```

```python
regions_data['srednie zaludnienie kraju'] = regions_data['peoples']/regions_data['countries']
regions_data
```

```python
# sortowanie od najbardziej zaludnionych regionów
regions_data.sort_values(by='srednie zaludnienie kraju', ascending = False)
```
